```python
import pandas as pd
import matplotlib.pyplot as plt
import string
import re

input_file = "suicide_detection_lowercased.csv"
df = pd.read_csv(input_file)

print("Dataset loaded for punctuation removal")

def count_punctuation(text):
    return sum(1 for c in str(text) if c in string.punctuation)

punct_before = df["text"].apply(count_punctuation).sum()



def remove_punctuation(text):
    return re.sub(rf"[{string.punctuation}]", "", text)

df["text"] = df["text"].apply(remove_punctuation)


punct_after = df["text"].apply(count_punctuation).sum()


plt.figure(figsize=(6,4))
plt.bar(
    ["Before Punctuation Removal", "After Punctuation Removal"],
    [punct_before, punct_after]
)
plt.title("Punctuation Count Before vs After Removal")
plt.ylabel("Total Punctuation Characters")
plt.xlabel("Stage")
plt.show()


output_file = "suicide_detection_punctuation_removed.csv"
df.to_csv(output_file, index=False)

print("Punctuation-removed dataset saved as:", output_file)
```
