```python
# Step 1- Lowercasing

import pandas as pd
import matplotlib.pyplot as plt

csv_file_name = "Suicide_Detection.csv"
df = pd.read_csv(csv_file_name)

print("Dataset loaded successfully")

def count_capitals(text):
    return sum(1 for c in str(text) if c.isupper())

capital_before = df["text"].apply(count_capitals).sum()

df["text"] = df["text"].str.lower()

capital_after = df["text"].apply(count_capitals).sum()


plt.figure(figsize=(6,4))
plt.bar(
    ["Before Lowercasing", "After Lowercasing"],
    [capital_before, capital_after]
)
plt.title("Capital Letter Count Before vs After Lowercasing")
plt.ylabel("Total Capital Letters")
plt.xlabel("Stage")
plt.show()


output_file = "suicide_detection_lowercased.csv"
df.to_csv(output_file, index=False)

print("Lowercased dataset saved as:", output_file)
```
